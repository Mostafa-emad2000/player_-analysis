```python
import numpy as np
import pandas as pd
```

```python
fb = pd.read_csv("players_20.csv")
```

```python
top_overall = fb.sort_values(by="overall",ascending=False).head(5)
print(top_overall)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
print (top_overall)
```

```python
player_information = ["attacking_finishing","attacking_short_passing","power_shot_power","power_long_shots","skill_long_passing","mentality_interceptions",
                      "mentality_vision","passing","skill_dribbling","dribbling","skill_long_passing","pace","power_strength","power_stamina",
                      "physic","movement_reactions","skill_moves","short_name"]
```

```python
df = pd.DataFrame(top_overall)
date_player = df[player_information]
print (date_player)
```

```python
print(date_player.isnull().sum())
```

```python
date_player["passing"].fillna(date_player["passing"].median(), inplace = True)
```

```python
date_player["dribbling"].fillna(date_player["dribbling"].median(), inplace = True)
```

```python
date_player["pace"].fillna(date_player["pace"].median(), inplace = True)
```

```python
date_player["physic"].fillna(date_player["physic"].median(), inplace = True)
```
